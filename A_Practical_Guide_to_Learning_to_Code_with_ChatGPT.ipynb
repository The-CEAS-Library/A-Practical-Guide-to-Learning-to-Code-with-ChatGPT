## **What are Large Language Models?** 

Large Language Models are AI models trained on vast amounts of text data from books, newspapers, websites, and more to understand* and generate human-like text. They are nueral networks with billions of parameters - They are not a database. At their core, they are built to predict the next word in a sentence, but with increased scale in training data and computer power, intelligence, reasoning skills, and other properties have emerged. The underlying technical details about LLMs are not the focus here, as this workshop aims to be practical in what we can do with LLMs as new tools. ## **Tips & Tricks for** ***Prompting*** 

The tips and tricks discussed below are not always necessary, but can be useful to employ when the model is not behaving like you want it to. As these models improve with time, the tricks below may become less important as the model learns to be equally helpful with minimal prompting. Think of these tips as tools in a tool box that you can pick and choose from in order to direct the model's outputs in a direction that is more productive and useful in aligning to your goals. You'll see how we selectively employ different tips in the examples below.


### **Set the Tone**

These models are essentially compressed versions of the internet, trained to interact or dialogue in ways we expect. This means it has all the best parts of the collective expertise online and the worst parts of the dark corners of the internet baked into its archetecture. As Cowen and Tabarrok put it "think of the AI as a 'collective
unconscious of humanity,' ala Carl Jung (who is Jung? Ask GPT!). You are trying to tap into that collective unconsciousness at a point that is interesting and in furtherance of your goals." 

This is important because the AI doesn't know where you are coming from, what your goals are, or the setting the conversation is taking place in. We have to tell it to be professional, to answer intelligently, or to be casual. 

To demonstrate, we can prompt ChatGPT with the silly, yet illustrative example:

```
Explain If and Else blocks in python using 
Tik Tok & Gen Z slang. 
```

Let's set a more professional tone with the model. 

```
That was fun, but I need to learn python for my job
requirements. Can you please explain the IF and Else
blocks in python in a professional way? 
```

 

### **Persona Adoption**

If you have explored the realm of "prompt engineering", you are likely aware of the power of persona adoption. This is sometimes phrased as "Ask for answers in the voice of experts". This helps direct the model's attention towards the smarter areas of its potential answer-space. This can also be accomplished by using a lot of intelligent sounding words in your prompt. Smarter prompts beget smarter answers.

In the case of using GPT4 as a coding tutor to learn python, something as simple as the following prompt is highly effective at setting a helpful tone for the interaction. 

---

```
You are going to take on the role of an excellent python coding 
tutor. As my coding tutor, please help me understand [insert coding concept]. 
```
Variations include


```
I want you to act as a code explainer in [language]...
```


```
I want you to act as a code debugger in [language]...
```


```

```

### **Be Specific** 

As discussed previously, you need to bring the context to the model. The model does not know what problem you are trying to solve and it is imperitive that you are specific with your prompts in order to get the most useful responses. This can mean going into a lot of detail about your specific problem and even the exact format you want the solution presented in. 

In order to get the most out of this tool, it is critical you write clear and specific instructions. You will note this in a few examples below. 

### **Give the Model Time to "Think"**

When you prompt the model, it will do its best to show you exactly what it thinks you want. This could mean simply providing an answer to a question or the fix to your code. When these questions involve multiple lines of reasoning, or many vertical logical steps in order to get to the correct conclusion, the model is prone to hallucinate and provide incorrect answers. 

This can be mitigated by giving the model time to think. In other words, we can create better prompts by implementing a few tactics. These include
*   Ask the model to show its steps or explain its reasoning. 

*   Outline the necessary steps to accomplish the task. 
*   Instruct the model to work out its own solution before it comes to a conclusion

 Instead of asking for too much at once, if we break up the ask into smaller individual tasks, the model is more likely to get each task correct. Let's look at the examples provided by Isa Fulford and Andrew Ng in their course - Prompt Engineering for Developers which demonstrate these tactics:

### **The Feynman Technique AKA ELI5**

Richard Feynman was known for his ability to explain complex physics concepts in simple terms. He emphasized the difference between understanding something and knowing the name of something as he was always after the deeper understanding. We can treat ChatGPT like our own personal "simplifier" by having it explain difficult coding concepts without jargon and complexity. 

When first encountering a new term in coding, it can be helpful to ask ChatGPT to explain it in a way a child could understand. This can yield remarkably powerful results for learning and understanding. 

To demonstrate, here is a response I got from ChatGPT4 to the prompt: `Explain object-oriented programming and user defined classes in a way a child could understand. `


```
GPT4: Sure! Imagine you have a box of LEGO bricks. 
Each brick is like a small piece of code that you can use to build something bigger. 
Object-oriented programming (OOP) is a way of organizing and 
building code using "bricks" called classes.  

A class is like a blueprint or a recipe that describes how to create something.
It tells you what pieces you need and how they should fit together. 
For example, a class might describe how to make a toy car, a toy airplane, or a toy robot. 

A user-defined class is a blueprint that you create yourself, 
instead of using one from the box of LEGOs. 
Let's say you want to create a new kind of toy animal that no one has ever seen before. 
You would create a user-defined class to describe how to build your toy animal. 

Now, when you want to build a toy using a class, you create an "object." 
An object is like a finished toy built using the blueprint. 
You can create many objects using the same class, 
just like you can build many toy animals using the same blueprint. 

Each object you create can have its own features, like a different color or a unique name. 
These features are called "attributes." Objects can also do things, 
like move, make noise, or light up. The actions an object can do are called "methods." 

In summary, object-oriented programming is a way to build code like you build with LEGO bricks. 
You create blueprints (classes) to describe how to make things (objects), 
and then you can use these objects to build bigger and more exciting creations! 
```

### **Drilling Down vs Clean Slate**

The power of an AI that you interact with through diologue should not be understated. Often times the model knows more about a topic than it states in an initial response, and through dialogue and drilling down on a concept, you can learn more and get better, more helpful explanations. 

On the other hand, sometimes the model gets confused, doesn't value what you think it should value, or simply "goes off the rails". This can happen for a number of reasons. Sometimes the best path forward is to create a new "context window" or chat session and start over. Chat history in one window does not immediately inform the model in another window. Every window is a blank slate with the current version of the model. Start over with setting the tone, giving a persona, and explaining your problem a different way to help the model better understand what you are after. It's worth noting that once a persona has been adopted and the tone has been set, it does not need to be reestablished within the same context window unless the model "forgets".
